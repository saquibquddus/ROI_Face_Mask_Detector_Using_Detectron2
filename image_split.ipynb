```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
im = cv.imread("sample_images/image6.png")
```

```python

im = cv.cvtColor(im, cv.COLOR_RGB2BGR)
```

```python
plt.imshow(im)
```

```python
length,width=im.shape[0:2]
length,width
```

```python
rect=im.copy()
rect=cv.rectangle(rect,(0,0),(int(width),int(length)),(0,0,0),3)
plt.imshow(rect)

```

```python
rec=im.copy()
plt.imshow(cv.rectangle(rec,(384,0),(510,128),(0,0,0),3))
```
